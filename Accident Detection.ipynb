```python
from keras.models import model_from_json
import numpy as np
import cv2
import os

class AccidentDetectionModel(object):
    class_nums = ['Accident', 'No Accident']

    def __init__(self, model_json_file, model_weights_file):
        # Load model from JSON file
        with open(model_json_file, 'r') as json_file:
            loaded_model_json = json_file.read()
            self.loaded_model = model_from_json(loaded_model_json)

        # Load weights into the new model
        self.loaded_model.load_weights(model_weights_file)
        self.loaded_model.make_predict_function()

    def predict_accident(self, img):
        self.preds = self.loaded_model.predict(img)
        return AccidentDetectionModel.class_nums[np.argmax(self.preds)], self.preds

# Load three models
model1 = AccidentDetectionModel('model.json', 'model.weights.h5')
model2 = AccidentDetectionModel('model2.json', 'model2.weights.h5')
model3 = AccidentDetectionModel('model3.json', 'model3.weights.h5')

# Initialize OpenCV video capture
video = cv2.VideoCapture("D:/Wafa/Documents/MINIPROJECT/Input/Video/Non Accident 1.mp4")  # Use 0 for default webcam

font = cv2.FONT_HERSHEY_SIMPLEX

while True:
    ret, frame = video.read()
    if not ret:
        break
    
    # Convert frame to RGB
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    # Resize frame
    roi = cv2.resize(gray_frame, (250, 250))
    
    # Predictions from all three models
    pred1, _ = model1.predict_accident(roi[np.newaxis, :, :])
    pred2, _ = model2.predict_accident(roi[np.newaxis, :, :])
    pred3, _ = model3.predict_accident(roi[np.newaxis, :, :])
    
    # Get majority prediction
    predictions = [pred1, pred2, pred3]
    majority_prediction = max(set(predictions), key=predictions.count)
    
    # Display prediction result on the video frame
    cv2.rectangle(frame, (0, 0), (280, 40), (0, 0, 0), -1)
    cv2.putText(frame, majority_prediction, (20, 30), font, 1, (255, 255, 0), 2)
    
    # Display the video frame
    cv2.imshow('Video', frame)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture object and close windows
video.release()
cv2.destroyAllWindows()
```
